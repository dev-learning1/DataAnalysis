###2. 밑 그림 그리기

```python
plt.rcParams['figure.figsize']=(5,5)

data = np.arange(1, 100)
plt.plot(data)  # 코랩기능 => 화면에 그래프 출력
plt.show()  # 그래프를 보여주는 코드
```

```python
data = np.arange(1, 51)
plt.plot(data)

data2 = np.arange(51, 101)
plt.plot(data2)

plt.show()
```

```python
# figure()는 새로운 그래프 canvas를 생성
data = np.arange(100, 200)
plt.plot(data)

data2 = np.arange(200, 301)
plt.figure()
plt.plot(data2)

plt.show()
```

```python
# 여러개의 plot를 그리는 방법
# subplot(row, column, index)
data = np.arange(100, 201)
plt.subplot(2, 1, 1)  #plt.subplot(211) 동일
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(2, 1, 2)  #plt.subplot(212) 동일
plt.plot(data2)

plt.show()
```

```python
# 1행 3열 짜리 그래프 3개를 작성
# 1행 1열(100~200), 1행 2열(200~300), 1행 3열(300~400)

data = np.arange(100, 201)
plt.subplot(131)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(132)
plt.plot(data2)

data3 = np.arange(300, 401)
plt.subplot(133)
plt.plot(data3)

plt.show()
```

###3. 주요 스타일 옵션

```python
plt.plot([1, 2, 3], [3, 6, 9])
plt.plot([1, 2, 3], [2, 4, 9])

# Title
plt.title('타이틀 테스트', fontsize=30)

# X, Y축 Label 설정
plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20, rotation=0)

plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(10)))

plt.title('범례 설정', fontsize=30)

# X, Y축 Label 설정
plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20, rotation=0)

# 눈금에 있는 글자
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# 범례(legend) 설정
plt.legend(['10*2', '10**2', 'log'], fontsize=15, loc='center right')

# x, y의 한계점(limit) 설정
plt.xlim(0, 10)
plt.ylim(0.5, 20)

plt.show()
```

###4. 스타일 세부 설정
[세부 document 확인](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

```python
plt.plot(np.arange(10), np.arange(10)*2, marker='o', markersize=5, linestyle='', color='b', alpha=0.3)  #alpha : 투명도(0~1)
plt.plot(np.arange(10), np.arange(10)**2, marker='v', markersize=5, linestyle='--', color='c', alpha=0.5)
plt.plot(np.arange(10), np.log(np.arange(10)), marker='*', markersize=13, linestyle='-.', color='y', alpha=0.7)

plt.title('마커 설정', fontsize=30)

# X, Y축 Label 설정
plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20, rotation=0)

# 눈금에 있는 글자
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# 그리드 설정
plt.grid()

plt.show()
```

```python
x = np.random.rand(50)  #0 ~ 1사이의 소수
y = np.random.rand(50)  
colors = np.arange(50)
area = x*y*1000

print(x)
print(y)
print(colors)
print(area)
```

```python
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
plt.figure(figsize=(12,6))

# cmap에 컬러를 설정하면 모든 그래프에 색상을 같게 설정

plt.subplot(131)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.1)
plt.title('alpha=0.1')

plt.subplot(132)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.5)
plt.title('alpha=0.5')

plt.subplot(133)
plt.scatter(x, y, s=area, cmap='blue', alpha=1.0)
plt.title('alpha=1.0')

plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Korea']
y = [67, 80, 50, 80, 10, 90]

plt.figure(figsize=(6, 3))
plt.bar(x, y, align='edge', alpha=0.7, color='red')
plt.xticks(x)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Korea']
y = [67, 80, 50, 80, 10, 90]

plt.figure(figsize=(6, 3))
# plt.bar(x, y, align='center', alpha=0.7, color='red')
plt.barh(x, y, align='center', alpha=0.7, color='green')
plt.yticks(x)
plt.xlabel('Number of Students')
plt.title('Subjects')

plt.show()
```

```python
# Line Plot
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin graph', fontsize=30)

plt.grid()

plt.show()
```

```python
# Pie Chart

labels = ['애플', '마이크로스프트', 'IBM', '나이키', '앤비디아', 'AT&T']
sizes = [20.5, 15.8, 10.5, 9, 7.6, 36.6]
explode = (0.3, 0, 0, 0, 0, 0)

patches, texts, autotexts = plt.pie(  #원 차트
    sizes,
    explode = explode, # 원 차트에서 해당 비율
    labels = labels,
    autopct = '%1.1f%%',
    shadow=True,
    startangle=90
)

plt.title('성공의 지름길', fontsize=30)

# label 텍스트에 대한 스타일 적용
for t in texts:
  t.set_fontsize(13)
  t.set_color('gray')

# pie 위에 텍스트에 대한 스타일 적용
for t in autotexts:
  t.set_color('white')
  t.set_fontsize(17)

plt.show()
```
